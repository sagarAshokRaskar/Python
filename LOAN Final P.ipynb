```python
import seaborn as sns
```

```python
import pandas as pd
```

```python
df=pd.read_csv('Downloads/Loan_dataset.csv')
```

```python
df.head(1)
```

## 1.What is the average current loan amount for each loan status?

```python
average_loan_amount = df.groupby('Loan Status')['Current Loan Amount'].mean()
print("Average current loan amount for each loan status:\n", average_loan_amount)
```

## 2.How does the credit score vary with the annual income?

```python
import matplotlib.pyplot as plt
plt.scatter(df['Annual Income'], df['Credit Score'])
plt.xlabel('Annual Income')
plt.ylabel('Credit Score')
plt.title('Credit Score vs Annual Income')
plt.show()
```

## 3.Is there a correlation between the number of open accounts and the current credit balance?

```python
df[['Number of Open Accounts','Number of Credit Problems']].corr()
```

```python
correlation_open_accounts_balance = df[['Number of Open Accounts', 'Current Credit Balance']].corr().iloc[0, 1]
print("Correlation between number of open accounts and current credit balance:", correlation_open_accounts_balance)
```

## 4.What is the distribution of credit scores across different home ownership types?

```python
credit_score_distribution = df.groupby('Home Ownership')['Credit Score'].describe()
print("Distribution of credit scores across different home ownership types:\n", credit_score_distribution)
```

## 5.How does the annual income differ for different purposes of loans?

```python
annual_income_by_purpose = df.groupby('Purpose')['Annual Income'].mean()
print("Annual income for different purposes of loans:\n", annual_income_by_purpose)
```
